```python
import requests
import json

def run_ollama(prompt: str, temperature=0.3):
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "llama3.1:8b",
        "prompt": prompt,
        "options": {"temperature": temperature}
    }

    res = requests.post(url, json=payload, stream=True)

    full_response = ""
    for line in res.iter_lines():
        if not line:
            continue
        data = json.loads(line.decode('utf-8'))
        token = data.get("response", "")
        print(f"token {token}")
        full_response += token
        if data.get("done"):
            break

    return full_response


print(run_ollama("What is LLM? Explain in one line.", temperature=0.3))
```
