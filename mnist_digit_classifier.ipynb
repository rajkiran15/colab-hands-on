```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from PIL import Image
import os

# --------------- Load and Prepare Data --------------- #
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize + reshape to vector of 784
x_train = x_train.reshape(-1, 784) / 255.0
x_test  = x_test.reshape(-1, 784) / 255.0

# --------------- Build Model --------------- #
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

# --------------- Compile & Train --------------- #
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nTraining model...")
model.fit(x_train, y_train, epochs=5, validation_split=0.2)

# --------------- Test Accuracy --------------- #
print("\nEvaluating on test data...")
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nğŸ”¥ Test Accuracy: {test_acc * 100:.2f}%")

# --------------- Predict on 5 new samples from test set --------------- #
print("\nPredicting first 5 test images...")
preds = model.predict(x_test[:5])

for i, p in enumerate(preds):
    print(f"Image {i+1} â†’ Predicted Digit:", p.argmax())

# --------------- Predict on 1 custom image from folder --------------- #
def predict_custom_image(img_path):
    # Load image and convert to grayscale
    img = Image.open(img_path).convert("L")

    # Resize only if not 28x28
    if img.size != (28, 28):
      img = img.resize((28, 28))

    img_array = np.array(img)

    # Flatten â†’ normalize â†’ reshape for model input
    img_array = img_array.reshape(1, 784) / 255.0

    pred = model.predict(img_array)
    return pred.argmax()

# Set your custom image path here
custom_folder = "digits"  # folder name
custom_image  = "digit_1.png"  # image file name
custom_path   = os.path.join(custom_folder, custom_image)

if os.path.exists(custom_path):
    result = predict_custom_image(custom_path)
    print("\nğŸ–¼ Custom Image Prediction:", result)
else:
    print("\nâš  No custom image found at:", custom_path)
```
